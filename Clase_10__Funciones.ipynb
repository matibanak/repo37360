```python
def funcion_multiplicar(*args):  #*args significa que uso la cantidad de datos que yo quiera con el *. no me limito a num1 y num2 solamente
  resultado=1
  for elemento in args:
    resultado*=elemento 
  return resultado
print(funcion_multiplicar(2,3,9))  #2 x 3 x 9
print(funcion_multiplicar(2,3,9,5,8,4,5,8))
```

```python
'''
Recursividad: proceso de repeticion en algo q se repite a si mismo
Tecnica para dividir un problema en partes mas pequeñas
Cuando se llama a si misma

'''

def cuenta_iterativa(numero):
  for i in range(1,numero+1):
    print(i)
cuenta_iterativa(8)
```

```python
def cuenta(numero):
  numero -=1

  if numero >0:
    cuenta(numero)
    print(f'---> {numero}')

cuenta(5)
```

```python
def factorialIterativo(n):
  if n<1:
    return 1
  else:
    resultado=1

    while(n>=1):
      resultado*=n
      n-=1
    return resultado

factorialIterativo(3)
```

```python
def factorial(n):

  if n==0 or n==1:
    return 1
  else:
    return (n*factorial(n-1))

factorial(3)
```
